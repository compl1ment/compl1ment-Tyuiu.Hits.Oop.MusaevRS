Вариант задания № 18 Описание задачи: Создать базовый класс Review в C#, который будет представлять отзывы о продуктах или услугах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма. Требования к базовому классу Review: • Атрибуты: ID отзыва (ReviewId), Текст отзыва (Text), Рейтинг (Rating). • Методы: o DisplayReview(): метод для отображения отзыва. o RateProduct(): метод для присвоения рейтинга продукту. o GetReviewDetails(): метод для получения деталей отзыва. Требования к производным классам:

ОтзывОбслуживания (ServiceReview): Должен содержать дополнительные атрибуты, такие как Дата посещения (VisitDate). Метод DisplayReview() должен быть переопределен для включения даты посещения в отображение отзыва.
ОтзывТовара (ProductReview): Должен содержать дополнительные атрибуты, такие как Идентификатор продукта (ProductId). Метод RateProduct() должен быть переопределен для связывания рейтинга с конкретным продуктом.
ОтзывУслуги (ServiceReview) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Время начала услуги (StartTime). Метод GetReviewDetails() должен быть переопределен для отображения времени начала услуги вместе с другими деталями отзыва.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
class Review // Создаем класс Review
{
    public int ReviewId{get;set;} // Создаем свойства с геттером и сеттером
    
    public string Text{get;set;}
    
    public int Rating{get;set;}
    

    public Review(int reviewId, string text, int rating)
    {
        ReviewId = reviewId;  
        Text = text;          
        Rating = rating;       
    }


    public virtual void DisplayReview() // Создаем публичные методы, которые даны в варианте
    {
        Console.WriteLine("Отзыв покупателя: " + Text);
    }
    public virtual void RateProduct()
    {
        Console.WriteLine("Рейтинг продукта: " + Rating);
    }
    public virtual void GetReviewDetails()
    {
        Console.WriteLine("ID покупателя: " + ReviewId);
    }
}

class ServiceMaintenance : Review // Создаем класс на основе Review. Это позволяет наследовать его свойства и методы
{
    private string startTime;
    public string StartTime
    {
        get {return startTime;}
        set {startTime = value;}

    }
    string NameofMaintenance{get;set;}

    public ServiceMaintenance(int reviewId, string text, int rating, string startTime,string nameofMaintenance, string visitDate): base(reviewId,text,rating)
    {
        NameofMaintenance = nameofMaintenance;
        VisitDate = visitDate;
    }

    private string visitDate;
    public string VisitDate
    {
        get {return visitDate;}
        set {visitDate = value;}

    }
    public override void DisplayReview() // Добавляем метод, необходимый для класса ServiceMaintenance, используя полиморфизм
    {
        Console.WriteLine($"Отзыв покупателя:{Text}. Услуга была оказана: {VisitDate}");
    }
    public virtual void DisplayNameOfMaintenance()
    {
        Console.WriteLine($"Услуга называется: {NameofMaintenance}");
    }
    
}

class ProductReview : ServiceMaintenance
{
    int ProductId{get;set;}
    
    public override void RateProduct()
    {
        Console.WriteLine($"Рейтинг продукта № {ProductId}:{Rating}"); // Добавляем метод, необходимый для класса ProductReview, используя полиморфизм
    }
    string Nameofproduct{get;set;}

    public ProductReview(int reviewId, string text, int rating, string startTime,string nameofproduct,int productId): base(reviewId,text,rating)
    {
        Nameofproduct = nameofproduct;
        ProductId = productId;
    }
    public virtual void NameProduct()
    {
        Console.WriteLine($"Название продукта: {Nameofproduct}"); // Добавляем метод, необходимый для класса ProductReview, используя полиморфизм
    }
}
class ServiceReview : Review
{
    string StartTime{get;set;}
    string NameofService{get;set;}   
    string VisitDate{get;set;}
    
    public ServiceReview(int reviewId, int rating,string text, string startTime,string nameofService,string visitDate): base(reviewId,text,rating)
    {
        NameofService = nameofService;
        VisitDate = visitDate;
        StartTime = startTime;
    }  
     
public ServiceReview(int reviewId, string text, int rating, string startTime,string nameofService,string visitDate): base(reviewId,text,rating)
{
    NameofService = nameofService;
    VisitDate = visitDate;
}

    public override void DisplayReview() 
    {
        Console.WriteLine($"Отзыв покупателя: {Text}.");
        Console.WriteLine($"Услуга была оказана: {VisitDate}. Время начала услуги: {StartTime}"); // Добавляем метод, необходимый для класса ServiceReview, используя полиморфизм
    }
    public virtual void GoodOrBadService()
    {
        if (Rating >= 4)
        {
            Console.WriteLine("Услуга оценивается покупателем высоко");
        }
        else
        {
            Console.WriteLine("Услуга оценивается покупателем низко");
        }
    }
}



ServiceReview serviceReview = new ServiceReview(4,6,"OK","14:10","Консультация у юриста","13.01.2024");
ServiceMaintenance serviceMaintenance = new ServiceMaintenance(3, "Доволен услугами тренера!", 6, "12:10","Личный тренер","13.10.2024");
ProductReview productReview = new ProductReview(11, "Планшет сломался при использовании!", 5, "","Samsung Galaxy S24 ultra",6);

Console.WriteLine("-------------------------------------------------------------------------------------"); 
productReview.DisplayReview();
productReview.RateProduct();
productReview.GetReviewDetails();
productReview.NameProduct();
Console.WriteLine("-------------------------------------------------------------------------------------"); 
serviceMaintenance.DisplayReview();
serviceMaintenance.RateProduct();
serviceMaintenance.GetReviewDetails();
Console.WriteLine("-------------------------------------------------------------------------------------"); 
serviceReview.DisplayReview();
serviceReview.RateProduct();
serviceReview.GetReviewDetails();
serviceReview.GoodOrBadService();
Console.WriteLine("-------------------------------------------------------------------------------------");
```
